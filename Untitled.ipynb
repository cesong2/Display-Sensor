```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
errors = pd.read_csv('C:/Display-Sensor/CGKim/archive/PdM_errors.csv')
failures = pd.read_csv('C:/Display-Sensor/CGKim/archive/PdM_failures.csv')
machines = pd.read_csv('C:/Display-Sensor/CGKim/archive/PdM_machines.csv')
maint = pd.read_csv('C:/Display-Sensor/CGKim/archive/PdM_maint.csv')
telemetry = pd.read_csv('C:/Display-Sensor/CGKim/archive/PdM_telemetry.csv')
```

```python
errors['datetime'] = pd.to_datetime(errors['datetime'])
failures['datetime'] = pd.to_datetime(failures['datetime'])
maint['datetime'] = pd.to_datetime(maint['datetime'])
telemetry['datetime'] = pd.to_datetime(telemetry['datetime'])
```

```python
errors.set_index('datetime',inplace=True)
failures.set_index('datetime',inplace=True)
machines.set_index('machineID',inplace=True)
maint.set_index('datetime',inplace=True)
telemetry.set_index('datetime',inplace=True)
```

```python
errors_pivot_table = errors.pivot_table(index=['machineID'], columns=['errorID'], aggfunc='size')
errors_pivot_table.fillna(0, inplace=True)
sns.heatmap(errors_pivot_table)
```

```python
failures_pivot_table = failures.pivot_table(index=['machineID'], columns=['failure'], aggfunc='size')
failures_pivot_table.fillna(0, inplace=True)
sns.heatmap(failures_pivot_table)
```

```python
maint_pivot_table = maint.pivot_table(index=['machineID'], columns=['comp'], aggfunc='size')
maint_pivot_table.fillna(0, inplace=True)
sns.heatmap(maint_pivot_table)
```
