```python
# 각 요인별 이상치 데이터 추출하는 함수 정의
def outlier(df, column):
    Q1 = np.quantile(df[column], 0.25)
    Q3 = np.quantile(df[column], 0.75)
    IQR = Q3 - Q1
    minimum = Q1 - (IQR * 1.5)
    maximum = Q3 + (IQR * 1.5)

    machineID = []
    value = []
    for i in range(len(df[column])):
        if df[column][i]>=maximum or df[column][i]<=minimum:
            machineID.append(df['기계ID'][i])
            value.append(df[column][i])
    return pd.DataFrame({'기계ID': machineID, column: value})
```

```python
# 각 요인별 이상치 데이터 추출
volt_outlier = outlier(telemetry_df, '전압')
rotate_outlier = outlier(telemetry_df, '회전')
pressure_outlier = outlier(telemetry_df, '압력')
vibration_outlier = outlier(telemetry_df, '진동')

# 이상치 데이터에서 기계별로 groupby, 각 이상치가 몇번씩 측정되었는지 count
volt_out_df = pd.DataFrame(volt_outlier.groupby('기계ID')['전압'].count())
rotate_out_df = pd.DataFrame(rotate_outlier.groupby('기계ID')['회전'].count())
pressure_out_df = pd.DataFrame(pressure_outlier.groupby('기계ID')['압력'].count())
vibration_out_df = pd.DataFrame(vibration_outlier.groupby('기계ID')['진동'].count())

volt_out_df.rename(columns={'전압':'volt_outliers'},inplace=True)
rotate_out_df.rename(columns={'회전':'rotate_outliers'},inplace=True)
pressure_out_df.rename(columns={'압력':'pressure_outliers'},inplace=True)
vibration_out_df.rename(columns={'진동':'vibration_outliers'},inplace=True)

# 기계별 고장횟수 & 이상치데이터 합치기
machines_outliers = pd.concat([machines_failures, volt_out_df, rotate_out_df, pressure_out_df, vibration_out_df], axis=1)
machines_outliers.rename(columns={'failure':'고장', 'volt_outliers':'전압_이상치', 'rotate_outliers':'회전_이상치', 'pressure_outliers':'압력_이상치', 'vibration_outliers':'진동_이상치', },inplace=True)
```

```python
# 기계별 고장 횟수와 이상치별 측정 횟수 상관관계 찾기 (피어슨 사용)
machines_outliers.corr(method = "pearson")
```
